## Part 1: Data Preparation and TF-IDF Analysis

### 1.1 Text Preprocessing


```python
def preprocess_text(text):

    text = text.lower()
    text = re.sub(r'[^а-яё\s]', '', text)

    tokens = word_tokenize(text, language='russian')

    # Remove stopwords and lemmatize
    processed_tokens = []
    for token in tokens:
        if token not in stop_words and len(token) > 2:

            lemma = morph.parse(token)[0].normal_form
            processed_tokens.append(lemma)

    return ' '.join(processed_tokens)


processed_songs = [preprocess_text(song) for song in songs]
```

### 1.2 TF-IDF Implementation


```python
tfidf_vectorizer = TfidfVectorizer(
    max_features=100,
    min_df=2,
    max_df=0.8
)

tfidf_matrix = tfidf_vectorizer.fit_transform(processed_songs)

feature_names = tfidf_vectorizer.get_feature_names_out()

tfidf_matrix, feature_names = filter_redundant_features(
    feature_names,
    tfidf_matrix,
    corr_threshold=0.95
)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    index=[f'Song {i+1}' for i in range(len(songs))],
    columns=feature_names
)

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
print(f"\nTop 20 features (words/phrases):")
print(feature_names[:20])
```

```python
def filter_redundant_features(feature_names, tfidf_matrix, corr_threshold=0.95):
    corr = np.corrcoef(tfidf_matrix.toarray(), rowvar=False)
    to_remove = set()

    for i in range(len(corr)):
        for j in range(i+1, len(corr)):
            if corr[i, j] > corr_threshold:
                to_remove.add(j)

    mask = [i for i in range(len(feature_names)) if i not in to_remove]

    new_matrix = tfidf_matrix[:, mask]
    new_features = feature_names[mask]

    return new_matrix, new_features
```

```python
top_features_per_song = {}

for idx, song_name in enumerate(tfidf_df.index):
    top_features = tfidf_df.loc[song_name].nlargest(15)
    top_features_per_song[song_name] = top_features

    print(f"\n{'='*60}")
    print(f"Top 15 TF-IDF features for {song_name}:")
    print(f"{'='*60}")
    for feature, score in top_features.items():
        print(f"{feature:30s} : {score:.4f}")

summary_data = []
for song_name, features in top_features_per_song.items():
    summary_data.append({
        'Song': song_name,
        'Top Feature': features.index[0],
        'Top Score': features.iloc[0],
        'Mean Score': features.mean(),
        'Std Score': features.std()
    })

summary_df = pd.DataFrame(summary_data)
print("\n" + "="*60)
print("Summary Statistics:")
print("="*60)
print(summary_df.to_string(index=False))
```

## Part 2: Comparison of Vectorization Methods

We will compare TF-IDF with:
1. Count Vectorizer
2. Word2Vec
3. Doc2Vec


```python
tokenized_songs = [song.split() for song in processed_songs]

print("="*60)
print("1. COUNT VECTORIZER")
print("="*60)
start_time = time.time()
count_vectorizer = CountVectorizer(max_features=100, ngram_range=(1, 2))
count_matrix = count_vectorizer.fit_transform(processed_songs)
count_time = time.time() - start_time
print(f"Computation time: {count_time:.4f} seconds")
print(f"Matrix shape: {count_matrix.shape}")
print(f"Non-zero elements: {count_matrix.nnz}")
print(f"Sparsity: {(1 - count_matrix.nnz / (count_matrix.shape[0] * count_matrix.shape[1])) * 100:.2f}%")
```

```python
print("\n" + "="*60)
print("2. WORD2VEC")
print("="*60)
start_time = time.time()
word2vec_model = Word2Vec(
    sentences=tokenized_songs,
    window=5,
    min_count=1,
    workers=1,
    sg=0
)
word2vec_time = time.time() - start_time
print(f"Computation time: {word2vec_time:.4f} seconds")
print(f"Vocabulary size: {len(word2vec_model.wv.key_to_index)}")
print(f"Vector dimensions: {word2vec_model.vector_size}")

def get_doc_vector_word2vec(model, tokens):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    return np.zeros(model.vector_size)

word2vec_doc_vectors = np.array([get_doc_vector_word2vec(word2vec_model, tokens)
                                  for tokens in tokenized_songs])
print(f"Document vectors shape: {word2vec_doc_vectors.shape}")
```

```python
print("\n" + "="*60)
print("3. DOC2VEC")
print("="*60)
start_time = time.time()
tagged_docs = [TaggedDocument(words=tokens, tags=[f'song_{i}'])
               for i, tokens in enumerate(tokenized_songs)]

doc2vec_model = Doc2Vec(
    documents=tagged_docs,
    vector_size=100,
    window=5,
    min_count=1,
    workers=1,
    epochs=50
)
doc2vec_time = time.time() - start_time
print(f"Computation time: {doc2vec_time:.4f} seconds")
print(f"Vocabulary size: {len(doc2vec_model.wv.key_to_index)}")
print(f"Vector dimensions: {doc2vec_model.vector_size}")

doc2vec_doc_vectors = np.array([doc2vec_model.dv[f'song_{i}']
                                for i in range(len(tokenized_songs))])
print(f"Document vectors shape: {doc2vec_doc_vectors.shape}")
```

```python
all_text = ' '.join(processed_songs)
all_tokens = all_text.split()

from collections import Counter

word_freq = Counter(all_tokens)
top_10_words = word_freq.most_common(10)

print("="*60)
print("TOP 10 MOST FREQUENTLY OCCURRING WORDS")
print("="*60)
for rank, (word, count) in enumerate(top_10_words, 1):
    print(f"{rank:2d}. {word:20s} : {count:3d} occurrences")

top_words_df = pd.DataFrame(top_10_words, columns=['Word', 'Frequency'])
```

```python
from collections import defaultdict

def extract_ngrams(tokens, n=2):
    """Extract n-grams from tokenized text"""
    ngrams = []
    for i in range(len(tokens) - n + 1):
        ngram = ' '.join(tokens[i:i+n])
        ngrams.append(ngram)
    return ngrams

all_bigrams = extract_ngrams(all_tokens, n=2)
bigram_freq = Counter(all_bigrams)
top_10_bigrams = bigram_freq.most_common(10)

print("\n" + "="*60)
print("TOP 10 MOST FREQUENTLY OCCURRING WORD COMBINATIONS (BIGRAMS)")
print("="*60)
for rank, (bigram, count) in enumerate(top_10_bigrams, 1):
    print(f"{rank:2d}. {bigram:40s} : {count:3d} occurrences")

top_bigrams_df = pd.DataFrame(top_10_bigrams, columns=['Bigram', 'Frequency'])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].barh(range(len(top_words_df)), top_words_df['Frequency'],
             color='#FF6B6B', alpha=0.8)
axes[0].set_yticks(range(len(top_words_df)))
axes[0].set_yticklabels(top_words_df['Word'], fontsize=11)
axes[0].set_xlabel('Frequency', fontsize=12, fontweight='bold')
axes[0].set_title('Top 10 Most Frequent Words', fontsize=14, fontweight='bold')
axes[0].invert_yaxis()
axes[0].grid(True, alpha=0.3, axis='x')

for i, v in enumerate(top_words_df['Frequency']):
    axes[0].text(v + 0.5, i, str(v), va='center', fontweight='bold')

axes[1].barh(range(len(top_bigrams_df)), top_bigrams_df['Frequency'],
             color='#4ECDC4', alpha=0.8)
axes[1].set_yticks(range(len(top_bigrams_df)))
axes[1].set_yticklabels(top_bigrams_df['Bigram'], fontsize=10)
axes[1].set_xlabel('Frequency', fontsize=12, fontweight='bold')
axes[1].set_title('Top 10 Most Frequent Word Combinations (Bigrams)',
                  fontsize=14, fontweight='bold')
axes[1].invert_yaxis()
axes[1].grid(True, alpha=0.3, axis='x')

for i, v in enumerate(top_bigrams_df['Frequency']):
    axes[1].text(v + 0.5, i, str(v), va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

### 3.3 t-SNE Visualization


```python
print("Computing t-SNE visualization...")
tsne = TSNE(n_components=2, random_state=42, perplexity=2, n_iter=1000)
tfidf_tsne = tsne.fit_transform(tfidf_matrix.toarray())

fig, ax = plt.subplots(figsize=(10, 8))

colors_tsne = ['#FF6B6B', '#4ECDC4', '#45B7D1']
for i, (x, y) in enumerate(tfidf_tsne):
    ax.scatter(x, y, s=300, c=colors_tsne[i], alpha=0.7,
               edgecolors='black', linewidth=2)
    ax.annotate(f'Song {i+1}', (x, y), fontsize=14, fontweight='bold',
                ha='center', va='center')

ax.set_title('t-SNE Visualization of Songs (TF-IDF Vectors)',
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('t-SNE Component 1', fontsize=12)
ax.set_ylabel('t-SNE Component 2', fontsize=12)
ax.grid(True, alpha=0.3)

distances = []
for i in range(3):
    for j in range(i+1, 3):
        dist = np.sqrt(np.sum((tfidf_tsne[i] - tfidf_tsne[j])**2))
        distances.append((f'Song {i+1} - Song {j+1}', dist))

print("\nEuclidean distances in t-SNE space:")
for pair, dist in distances:
    print(f"  {pair}: {dist:.3f}")

plt.tight_layout()
plt.show()
```
